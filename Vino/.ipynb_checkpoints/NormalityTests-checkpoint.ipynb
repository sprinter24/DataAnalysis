### Проверка на нормальность

```python
plt.figure(figsize = (12,6)) #общий размер фигуры: 12 на 6
plt.title("age") 
plt.hist(data['age'], bins=15) #bins = количество столбцов гистограммы
plt.legend(loc = 4)
# legend -  расположение подписей графиков (номер четверти)
pass

```

### Q-Q plot или рисование на нормальной бумаге

Если данные имеют нормальное распределение (хоть с какими-то параметрами), то точки на нормальной бумаге должны лежать на прямой

```python
plt.figure(figsize = (20,10))
fig = sm.qqplot(data['age'], stats.norm)
plt.show()
pass
```

### Shapiro-Wilk

Выкинем пустые значения

```python
age = data['age'].dropna()
age[:30]
```

```python
stats.shapiro(age[:20])
```

pvalue >> 0,05  - и вообще большое - Нулевую гипотезу о нормальности - ПРИНИМАЕМ!!!

pvalue=0.4801372289657593 - значит данные хорошо согласуются с нулевой гипотезой

```python
stats.shapiro(age[-10:])
```

pvalue < 0,05  - Нулевую гипотезу о нормальности - ОТВЕРГАЕМ!!!

```python
stats.shapiro(age[-100:])
```

```python
stats.shapiro(age)
```

ТУТ гипотезу о нормальности ОТВЕРГАЕМ с большой уверенностью

### Liliefors test - проверка нормальности на основе теста Колмогорова

```python
sm.stats.diagnostic.lilliefors(age, dist='norm', pvalmethod='table')
```

НОРМАЛЬНОСТЬ ОТВЕРГАЕМ

```python
sm.stats.diagnostic.lilliefors(age[:20], dist='norm', pvalmethod='table')
```

НАЧАЛО ВЫБОРКИ - НОРМАЛЬНОЕ!!!

### Jarque-Bera test - тест Харке-Бера, основанный на ассиметрии и эксцессе

```python
stats.jarque_bera(age)
```

НОРМАЛЬНОСТЬ ОТВЕРГАЕМ

```python
stats.jarque_bera(age[:20])
```

НАЧАЛО ВЫБОРКИ - НОРМАЛЬНОЕ!!!
